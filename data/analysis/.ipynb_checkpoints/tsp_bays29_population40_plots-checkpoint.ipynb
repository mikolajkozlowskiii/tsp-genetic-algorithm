```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_json('bays29_population40.json')
```

```python
df.shape
```

```python
df.columns
```

```python
pd.DataFrame(df['tspOutputs'].tolist()).columns
```

```python
plt.figure(figsize=(12, 8))

for _, row in df.iterrows():
    mutation_type = row['mutationType']
    crossover_type = row['crossoverType']
    selection_type = row['selectionType']
    tsp_outputs = row['tspOutputs']
    
    all_generations = []
    for output in tsp_outputs:
        generations = output['generations']
        all_generations.extend(generations)
    
    gen_df = pd.DataFrame(all_generations)
    gen_df['iteration'] = gen_df['iteration'].astype(int)
    gen_df['result'] = gen_df['result'].astype(float)
    
    average_results = gen_df.groupby("iteration")["result"].mean().reset_index()
    
    label = f"Crossover: {crossover_type}, Mutation: {mutation_type}, Selection: {selection_type}"
    plt.plot(average_results["iteration"], average_results["result"], marker='o', label=label)

plt.title("Average TSP Results per Iteration for Different Parameters, population 40")
plt.xlabel("Iteration")
plt.ylabel("Average Result")
plt.grid(True)
plt.legend(loc="upper right", fontsize="small")
plt.tight_layout()

plt.show()
```

```python
selection_groups = df.groupby("selectionType")

fig, axes = plt.subplots(len(selection_groups), 1, figsize=(12, 6 * len(selection_groups)), sharex=True)


if len(selection_groups) == 1:
    axes = [axes]

for ax, (selection_type, group) in zip(axes, selection_groups):
    for _, row in group.iterrows():
        tsp_outputs = row['tspOutputs']
        all_generations = []
        for output in tsp_outputs:
            generations = output['generations']
            all_generations.extend(generations)
        
        gen_df = pd.DataFrame(all_generations)
        gen_df['iteration'] = gen_df['iteration'].astype(int)
        gen_df['result'] = gen_df['result'].astype(float)

        average_results = gen_df.groupby("iteration")["result"].mean().reset_index()

        label = f"Crossover: {row['crossoverType']}, Mutation: {row['mutationType']}"
        ax.plot(average_results["iteration"], average_results["result"], marker='o', label=label)

    ax.set_title(f"Selection Type: {selection_type}")
    ax.set_xlabel("Iteration")
    ax.set_ylabel("Average Result")
    ax.legend(loc="upper right", fontsize="small")
    ax.grid(True)

plt.tight_layout()

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import json

file_path = "all-tests-7.json"
with open(file_path, 'r') as file:
    data = json.load(file)

df = pd.json_normalize(
    data,
    record_path='tspOutputs',  
    meta=['mutationType', 'crossoverType', 'selectionType'],
    errors='ignore'
)

df['mutationType'] = df['mutationType'].astype(str)
df['crossoverType'] = df['crossoverType'].astype(str)
df['selectionType'] = df['selectionType'].astype(str)

df['combination'] = df['mutationType'] + "-" + df['crossoverType'] + "-" + df['selectionType']

selection_groups = df.groupby('selectionType')

for selection_type, group in selection_groups:
    plt.figure(figsize=(10, 6))
    group.boxplot(column='distance', by='combination', grid=False, rot=90)
    plt.title(f"Boxplot for TSP Results with Selection Type: {selection_type}")
    plt.suptitle("")  # Remove default title added by `boxplot`
    plt.xlabel("Operator Combination")
    plt.ylabel("Distance Result")
    plt.tight_layout()
    plt.show()
```
